```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_phishing.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.status.value_counts()
```

```python
df1 = df.drop("url" , axis = "columns")
df1.head()
```

```python
#df2 = df1.drop(df1.loc[: , "nb_hyperlinks" : "page_rank"].columns , axis = "columns")
#df2.head()
```

```python
dummies = pd.get_dummies(df1.status)
dummies.head()
merged = pd.concat([df1,dummies] , axis='columns')
merged.head()
```

```python
df_final = merged.drop("status", axis = "columns")
df_final.head()
```

```python
#diffrentiate categorical column and numeric column
```

```python
cat_cols = {}
for i in df_final.iloc[:].columns:
    cat_cols[i] = df_final[i].nunique()/df_final[i].count() < 0.002
    

num_col = []
cat_col = []

for col in cat_cols.keys():
    if(cat_cols[col] == False):
        num_col.append(col)
    else:
        cat_col.append(col)

print("Number of numeric columns:",len(num_col))
print(num_col)
print("\n")
print("Number of categorical columns:",len(cat_col))
print(cat_col)
```

```python
#implement correlation matrix for features
```

```python
import seaborn as sns
#from matplotlib.patches import Patch
```

```python
cor_matrix = df_final[num_col].corr()

matrix = plt.figure(figsize=(12,12) , dpi=70)
mask = np.triu(np.ones_like(cor_matrix , dtype = bool))
sns.heatmap(cor_matrix , mask=mask , cmap = "YlGnBu" , robust=True , center=0 , square=True , linewidths=0.8)
plt.title("Correlation matrix of numeric features" , fontsize=12)
```

```python
cor_matrix = df_final[cat_col].corr()

matrix = plt.figure(figsize=(12,12) , dpi=70)
mask = np.triu(np.ones_like(cor_matrix , dtype = bool))
sns.heatmap(cor_matrix , mask=mask , cmap = "YlGnBu" , robust=True , center=0 , square=True , linewidths=0.8)
plt.title("Correlation matrix of categorical features" , fontsize=12)
```

```python
#how impactful one particular feature on the decision value based on the mean value
```

```python
df_mean1 = df_final.groupby("legitimate")[num_col].mean().reset_index().T
df_mean1.rename(columns ={0:"Phishing", 1:"Legitimate"} , inplace = True)
plt.rcParams["axes.facecolor"] = "w"
#graph = df_mean1.plot(kind="bar" , title = "Distribution of average value over legitimate value")
graph = df_mean1.plot(kind="bar" , title = "Distribution of average value over legitimate value" , figsize=(10,5) , legend=True , fontsize = 10)
graph.set_xlabel("Numerical features" , fontsize=12)
graph.set_ylabel("Average values" , fontsize=12)
plt.show()
```

```python
cols = list(df_final.columns)
```

```python
df_x = df_final[cols[:-2]]
df_y = df_final['legitimate']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

clr = RandomForestClassifier(n_estimators=20)



#from sklearn.svm import SVR
#svr = SVR(kernel = 'linear', C=1)
```

### Standardize data

```python
from sklearn.preprocessing import StandardScaler

sc= StandardScaler().fit(x_train)

x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
clr.fit(x_train_std, y_train)
```

```python
clr.score(x_train_std , y_train)
```

```python
clr.predict(x_test_std)
```

```python
clr.score(x_test_std , y_test)
```
